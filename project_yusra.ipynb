Price Analysis (NumPy Feature 1)
Identify the price column from the dataset.
Convert all values to numeric and handle missing data.
Calculate summary statistics: count, mean, median, standard deviation, min, and max.
Detect outliers using z-scores (values more than 3 standard deviations from the mean).

```python
import numpy as np
import pandas as pd

# Load CSV
df = pd.read_csv('Airbnb_Open_Data.csv', encoding='utf-8')

# 1) Find the price column
price_col = None
for col in df.columns:
    if 'price' in col.lower():
        price_col = col
        break

prices = df[price_col]

# Convert prices to float and handles missing
prices_clean = pd.to_numeric(prices, errors='coerce')
print(f"Using price column: {price_col}")
print("Missing after parse:", prices_clean.isna().sum(), "/", len(prices_clean))

# Summary statistics
count = prices_clean.count()
mean = np.mean(prices_clean)
median = np.median(prices_clean)
std = np.std(prices_clean)
min_price = np.min(prices_clean)
max_price = np.max(prices_clean)

print("\nPrice summary (cleaned):")
print("count:", count)
print("mean:", mean)
print("median:", median)
print("std:", std)
print("min:", min_price)
print("max:", max_price)

# Detect outliers (z-score > 3)
z_scores = (prices_clean - mean) / std
outliers = np.sum(np.abs(z_scores) > 3)
print("\nOutliers detected (z>|3|):", outliers)
```

Objective: Analyze which room types are most frequently reviewed by guests.
Extracted room type and reviews per month columns.
Converted reviews per month to numeric, ignoring missing values.
Used NumPy to calculate the average reviews per month for each room type.
output:
Hotel rooms get the highest average reviews per month (2.84).
Entire homes/apartments and private/shared rooms have lower review rates (~1.3–1.4).
This shows hotel rooms attract more frequent guest feedback, possibly due to higher turnover.

```python
import numpy as np

# Extract columns
room_types = df['room type'].values
reviews_per_month = df['reviews per month'].values

# Convert reviews per month to numeric, missing values become NaN
reviews_per_month = np.array(pd.to_numeric(reviews_per_month, errors='coerce'))

# Find unique room types
unique_rooms = np.unique(room_types)

print("Average reviews per month by room type:")
for room in unique_rooms:
    mask = room_types == room
    avg_reviews = np.nanmean(reviews_per_month[mask])  # ignores NaN
    print(f"{room}: {avg_reviews:.2f}")
```

This class defines a Host object for analyzing Airbnb host performance.This score helps identify hosts who are actively engaging with guests while managing their availability.

```python
class Host: 
    def __init__(self, host_id, host_name, reviews_per_month, number_of_reviews, availability_365, review_rate_number=None):
        self.host_id = host_id
        self.host_name = host_name
        self.reviews_per_month = reviews_per_month
        self.number_of_reviews = number_of_reviews
        self.availability_365 = availability_365
        self.review_rate_number = review_rate_number  #  used in subclasses

    def calculate_score(self):
        return (self.reviews_per_month * self.number_of_reviews) - self.availability_365

    def performance_summary(self):
        return f"{self.host_name} Score: {self.calculate_score()}"
```

extract some of the columns necessary for host performance analysis from the full datase.this creates a smaller DataFrame hosts_df that will be used to compute host performance scores.

```python
hosts_df = df[['host id',
               'host name',
               'reviews per month',
               'number of reviews',
               'availability 365',
               'review rate number'   
              ]]

hosts_df.head()

     
```

making sure each host has 1 row even if they have multiple listings

```python
grouped = hosts_df.groupby('host id').agg({
    'host name': 'first',
    'reviews per month': 'mean',
    'number of reviews': 'sum',
    'availability 365': 'mean',
    'review rate number': 'mean'   
}).reset_index()

print(grouped.head())

```

Before performing calculations, replace any missing (NaN) values with 0.This way the score calculation will not give errors or NaN results.By filling missing values with 0, hosts with incomplete data can still be included in the performance analysis

```python
# Fill missing values with 0
grouped['reviews per month'] = grouped['reviews per month'].fillna(0)
grouped['number of reviews'] = grouped['number of reviews'].fillna(0)
grouped['availability 365'] = grouped['availability 365'].fillna(0)
grouped['review rate number'] = grouped['review rate number'].fillna(0)   

```

convert each row of the grouped dataframe into a Host object and store them in hosts_list 

```python
hosts_list = []

for index, row in grouped.iterrows():
    host_obj = Host(
        host_id=row['host id'],
        host_name=row['host name'],
        reviews_per_month=row['reviews per month'],
        number_of_reviews=row['number of reviews'],
        availability_365=row['availability 365'],
        review_rate_number=row['review rate number']   
    )
    hosts_list.append(host_obj)

# Checks first 5 hosts
hosts_list[:5]
```

each Host object’s score is calculated using defined formula.Scores are stored in a dictionary with host names as keys.
Finally sort the hosts by score in descending order to identify the top performers.

```python
host_scores = {}

for host in hosts_list:
    score = host.calculate_score()
    host_scores[host.host_name] = score
# Sorting by score descending and display top 10
top_hosts = sorted(host_scores.items(), key=lambda x: x[1], reverse=True)
top_hosts[:10]

```

convert the host scores into a dataframe and sort it in descending order.This shows the top 10 hosts along with their performance scores.

```python
import pandas as pd
# Convert the host_scores dictionary to a DataFrame
host_scores_df = pd.DataFrame(list(host_scores.items()), columns=['Host Name', 'Score'])
# Sort by score descending
host_scores_df = host_scores_df.sort_values(by='Score', ascending=False)
# Display top 10 hosts by score
print(host_scores_df.head(10))


```

This code demonstrates inheritance and polymorphism.'HighRatedHost' and 'LowRatedHost' inherit from the base Host class.
Both subclasses override the performance_summary method (polymorphism) to display messages based on the host's rating. Host objects are created from the grouped dataset and their performance summaries are displayed.


```python
# base method to Host class
def performance_summary(self):
    return self.host_name + " Score: " + str(self.calculate_score())

# Subclass for high-rated hosts (inhritance+ polymorphism)
class HighRatedHost(Host):
    def __init__(self, host_id, host_name, reviews_per_month, number_of_reviews, availability_365, review_rate_number):
        super().__init__(host_id, host_name, reviews_per_month, number_of_reviews, availability_365)
        self.review_rate_number = review_rate_number

    # POLYMORPHISM: override method
    def performance_summary(self):
        return f"High Rated Host :{self.host_name} (Rating {self.review_rate_number}) - Score: {self.calculate_score()}"

# Subclass for low-rated hosts (INHERITANCE + POLYMORPHISM)
class LowRatedHost(Host):
    def __init__(self, host_id, host_name, reviews_per_month, number_of_reviews, availability_365, review_rate_number):
        super().__init__(host_id, host_name, reviews_per_month, number_of_reviews, availability_365)
        self.review_rate_number = review_rate_number
  # POLYMORPHISM: override method
    def performance_summary(self):
        return f"Low Rated Host : {self.host_name} (Rating {self.review_rate_number}) - Score: {self.calculate_score()}"

# Create objects for all hosts
hosts_objects = []
for index, row in grouped.iterrows():
    if row['review rate number'] >= 4:
        host_obj = HighRatedHost(
            host_id=row['host id'],
            host_name=row['host name'],
            reviews_per_month=row['reviews per month'],
            number_of_reviews=row['number of reviews'],
            availability_365=row['availability 365'],
            review_rate_number=row['review rate number']
        )
    else:
        host_obj = LowRatedHost(
            host_id=row['host id'],
            host_name=row['host name'],
            reviews_per_month=row['reviews per month'],
            number_of_reviews=row['number of reviews'],
            availability_365=row['availability 365'],
            review_rate_number=row['review rate number']
        )
    hosts_objects.append(host_obj)

# Display performance summaries for first 10 hosts
for host in hosts_objects[:10]:
    print(host.performance_summary())
```
